# Criando labels para a relev√¢ncia da vaga anunciada

```python
df = pd.read_csv("linkedin_jobs.csv")
```

```python
df.head()
```

```python
import pandas as pd

# Create a new column for relevance
df['RELEVANCIA'] = ''

# Counter for tracking the progress
counter = 0

# Loop through the DataFrame and prompt for relevance attribution
while counter < len(df):
    entry = df.iloc[counter]
    title = entry['TITULO']
    location = entry['LOCAL']
    description = entry['DESCRICAO']
    
    # Display the entry details
    print(f"Entry {counter + 1}:")
    print(f"Title: {title}")
    print(f"Location: {location}")
    print(f"Description: {description}")
    
    # Get the relevance attribution from the user
    relevance = input("Enter the relevance attribution (0, 1, or 2): ")
    
    # Validate the relevance input
    while relevance not in ['0', '1', '2']:
        print("Invalid relevance attribution. Please enter 0, 1, or 2.")
        relevance = input("Enter the relevance attribution (0, 1, or 2): ")
    
    # Assign the relevance attribution to the corresponding row
    df.at[counter, 'RELEVANCIA'] = relevance
    
    # Increment the counter
    counter += 1
    
    # Print a separator
    print('-' * 40)

# Save the classified DataFrame to a CSV file
df.to_csv('dados_classificados.csv', index=False)
```

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("dados_classificados.csv")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

# Combine the 'TITULO' and 'DESCRICAO' columns into a single column
data['TEXT'] = data['TITULO'] + ' ' + data['DESCRICAO']

# Split the data into a training set (80%) and a test set (20%)
train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)

# Initialize a TfidfVectorizer
vectorizer = TfidfVectorizer(sublinear_tf=True, encoding='utf-8', decode_error='ignore', stop_words='english')

# Fit and transform the 'TEXT' column of the training data
train_features = vectorizer.fit_transform(train_data['TEXT'])

# Transform the 'TEXT' column of the test data
test_features = vectorizer.transform(test_data['TEXT'])

# Extract the 'RELEVANCIA' columns of the training and test data
train_labels = train_data['RELEVANCIA']
test_labels = test_data['RELEVANCIA']
```

We've combined the 'TITULO' and 'DESCRICAO' columns into a single 'TEXT' column.
We've used the TF-IDF technique to convert the text data into a matrix of TF-IDF features.
We've split the data into a training set (80%) and a test set (20%).

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# Initialize a RandomForestClassifier
clf = RandomForestClassifier(random_state=42)

# Train the classifier
clf.fit(train_features, train_labels)

# Make predictions on the test data
test_preds = clf.predict(test_features)

# Evaluate the performance of the model
report = classification_report(test_labels, test_preds)

report
```

```python
from sklearn.model_selection import GridSearchCV

# Define the hyperparameters and their values
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

# Initialize a GridSearchCV
grid_search = GridSearchCV(clf, param_grid, cv=5, scoring='f1_macro', verbose=1, n_jobs=-1)

# Fit the GridSearchCV
grid_search.fit(train_features, train_labels)

# Get the best parameters and the best score
best_params = grid_search.best_params_
best_score = grid_search.best_score_

best_params, best_score
```

```python
# Initialize a RandomForestClassifier with the best parameters
clf_best = RandomForestClassifier(**best_params, random_state=42)

# Train the classifier
clf_best.fit(train_features, train_labels)

# Make predictions on the test data
test_preds_best = clf_best.predict(test_features)

# Evaluate the performance of the model
report_best = classification_report(test_labels, test_preds_best)

report_best
```

```python
# Initialize a RandomForestClassifier with the best parameters and class_weight='balanced'
clf_balanced = RandomForestClassifier(**best_params, class_weight='balanced', random_state=42)

# Train the classifier
clf_balanced.fit(train_features, train_labels)

# Make predictions on the test data
test_preds_balanced = clf_balanced.predict(test_features)

# Evaluate the performance of the model
report_balanced = classification_report(test_labels, test_preds_balanced)

report_balanced
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
import xgboost as xgb
from sklearn.metrics import classification_report

# Combine as classes 1 e 2 em uma √∫nica classe (classe 1)
data_dropped['RELEVANCIA'] = data_dropped['RELEVANCIA'].replace([2], [1])

# Divis√£o dos dados em conjunto de treinamento e conjunto de teste
train_data, test_data = train_test_split(data_dropped, test_size=0.2, random_state=42)

# Transforma as colunas 'TITULO' e 'DESCRICAO' em um √∫nico texto
train_data['TEXT'] = train_data['TITULO'] + ' ' + train_data['DESCRICAO']
test_data['TEXT'] = test_data['TITULO'] + ' ' + test_data['DESCRICAO']

# Combina as stopwords em uma √∫nica lista
stopwords_combined = list(stopwords_english) + list(stopwords_portuguese)

# Transforma o texto em features
vectorizer = TfidfVectorizer(stop_words=stopwords_combined)
train_features = vectorizer.fit_transform(train_data['TEXT'])
test_features = vectorizer.transform(test_data['TEXT'])

# Extrai as colunas 'RELEVANCIA' dos dados de treinamento e teste
train_labels = train_data['RELEVANCIA']
test_labels = test_data['RELEVANCIA']

# Calcula a raz√£o entre as classes majorit√°ria e minorit√°ria
ratio = len(train_labels[train_labels == 0]) / len(train_labels[train_labels == 1])

# Define os hiperpar√¢metros do XGBoost
params = {
    'scale_pos_weight': ratio,
    'subsample': 0.8,
    'max_delta_step': 1
}

# Inicializa o classificador XGBoost com os par√¢metros definidos
xgb_classifier = xgb.XGBClassifier(random_state=42, **params)

# Treina o classificador
xgb_classifier.fit(train_features, train_labels)

# Faz as previs√µes nos dados de teste
test_preds = xgb_classifier.predict(test_features)

# Avalia o desempenho do modelo
report = classification_report(test_labels, test_preds)

print(report)
```
